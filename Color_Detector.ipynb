```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('Example_image.png')
rgb_img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
```

```python
hsv = cv2.cvtColor(rgb_img,cv2.COLOR_RGB2HSV)
```

```python
#Yellow
lower_range_y = np.array([25,150,50])
upper_range_y = np.array([35,255,255])

#Red
lower_range_r = np.array([169,100,100])
upper_range_r = np.array([189,255,255])

#Purple
lower_range_p = np.array([130,141,164])
upper_range_p = np.array([160,141,165])

#Blue
lower_range_b = np.array([110,150,50])
upper_range_b = np.array([130,255,255])

#Green
lower_range_g = np.array([59,206,177])
upper_range_g = np.array([79,206,177])

#Orange 
lower_range_o = np.array([2,216,255])
upper_range_o = np.array([22,216,255])
```

```python
#Yellow
mask_y = cv2.inRange(hsv,lower_range_y,upper_range_y)

#Red
mask_r = cv2.inRange(hsv,lower_range_r,upper_range_r)

#Purple
mask_p = cv2.inRange(hsv,lower_range_p,upper_range_p)

#Blue
mask_b = cv2.inRange(hsv,lower_range_b,upper_range_b)

#Green
mask_g = cv2.inRange(hsv,lower_range_g,upper_range_g)

#Orange
mask_o = cv2.inRange(hsv,lower_range_o,upper_range_o)
```

```python
#Yellow
res = cv2.bitwise_and(rgb_img,rgb_img,mask = mask_y)
plt.imshow(res)
```

```python
#Red
res = cv2.bitwise_and(rgb_img,rgb_img,mask = mask_r)
plt.imshow(res)
```

```python
#Purple
res = cv2.bitwise_and(rgb_img,rgb_img,mask = mask_p)
plt.imshow(res)
```

```python
#Green
res = cv2.bitwise_and(rgb_img,rgb_img,mask = mask_g)
plt.imshow(res)
```

```python
#Blue
res = cv2.bitwise_and(rgb_img,rgb_img,mask = mask_b)
plt.imshow(res)
```

```python
#Orange
res = cv2.bitwise_and(rgb_img,rgb_img,mask = mask_o)
plt.imshow(res)
```
